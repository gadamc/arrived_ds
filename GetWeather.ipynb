```python
aprildays = [29,30]
maydays = range(1,31)
junedays = [1,2,3,4,5,6,7,8]
days = ['20160429', '20160430']
days += ['201605{:02}'.format(x) for x in maydays]
days += ['201606{:02}'.format(x) for x in junedays]
days
```

```python
import time
weather_data = {}
for day in days:
    f = requests.get('http://api.wunderground.com/api/ad2a68f2f4694ad3/history_{}/q/pws:IESVITOR2.json'.format(day))
    time.sleep(0.1)
    try:
        f.raise_for_status()
        weather_data[day] = f.json()
    except:
        print("failed for: {}".format(day))
```

### Write Daily Summary to CSV

```python
features = [k for k in parsed_json['history']['dailysummary'][0].keys() if k not in ('date','utcdate')]
```

```python
date_features = ['date_{}'.format(k) for k in parsed_json['history']['dailysummary'][0]['date'].keys()]
```

```python
date_features
```

```python
features
```

```python
import csv
with open('weather_dailysummary.csv', 'w', newline='') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=features + date_features)
    writer.writeheader()
    for day in days:
        ds = weather_data[day]['history']['dailysummary'][0]
        row = {k:ds[k] for k in features}
        row.update({k:ds['date'][k.split('_')[1]] for k in date_features})
        writer.writerow(row)
            
```
