```python
import pandas as pd
```

```python
filename = 'KaggleV2-May-2016_with_new_cols_may5_1338.csv'
no_show = pd.read_csv(filename, encoding='latin')
```

```python
weather_dailysummary = 'weather_dailysummary.csv'
weather = pd.read_csv(weather_dailysummary)
```

```python

weather['AppointmentDay'] = pd.to_datetime(weather['date_pretty'])
```
